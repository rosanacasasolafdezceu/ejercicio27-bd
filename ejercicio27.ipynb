```python
import pymysql

db = pymysql.connect(
    host="localhost",
    user="root",
    password="",
    database="prod_hipoteca"
)

cursor = db.cursor()

print("Conectado correctamente")
```

```python
cursor.execute("DROP TABLE IF EXISTS PROPIEDAD")
```

```python
#Voy a crear la tabla propiedad

cursor.execute("""
CREATE TABLE PROPIEDAD (
    id_propiedad INT PRIMARY KEY,
    direccion VARCHAR(100),
    metros INT,
    precio DECIMAL(10,2),
    dni_propietario VARCHAR(20)
)
""")

print("Tabla PROPIEDAD creada")
```

```python
#Inserto las cosas del propietario

cursor.execute("""
INSERT INTO PROPIEDAD VALUES (1, 'Calle Pez 8', 12, 28003, '1234578A')
""")

cursor.execute("""
INSERT INTO PROPIEDAD VALUES (2, 'Avenida de la Moncloa 8', 120, 11310, '1234578A')
""")

db.commit()

print("Datos insertados")
```

```python
#Compruebo si esos datos se han insertado bien

cursor.execute("SELECT * FROM PROPIEDAD")
print(cursor.fetchall())
```

```python
#Borro tabla por si existe como antes

cursor.execute("DROP TABLE IF EXISTS propiedad_copia")
```

```python
#Voy a crear la tabla copia

cursor.execute("""
CREATE TABLE propiedad_copia (
    id_propiedad INT PRIMARY KEY,
    direccion VARCHAR(100),
    metros INT,
    precio DECIMAL(10,2),
    dni_propietario VARCHAR(20)
)
""")

print("Tabla propiedad_copia creada")
```

```python
#Insertar usando select

cursor.execute("""
INSERT INTO propiedad_copia
SELECT * FROM PROPIEDAD
""")

db.commit()

print("Datos copiados")
```

```python
#Consultar el promedio (hay q calcularlo)

cursor.execute("SELECT AVG(precio) FROM PROPIEDAD")

resultado = cursor.fetchone()

print("Precio medio:", resultado[0])
```
